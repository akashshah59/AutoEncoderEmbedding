```python
import torch
import pickle
import matplotlib.pyplot as plt
```

```python
model = torch.load('auto_encoder.pkl')
```

```python
with open('train.pkl','rb') as f:
    x = torch.tensor(pickle.load(f)).double()

with open('test.pkl','rb') as f:
    y = pickle.load(f)
```

```python
embeddings = model.encoder(x)
```

```python
embeddings.shape
```

```python
y
```

```python
plt.figure(figsize=(20,10))
plt.title('Auto encoded Embeddings')
plt.scatter(embeddings[:,0].detach().numpy(),embeddings[:,1].detach().numpy(),c = y)
plt.show()
```
